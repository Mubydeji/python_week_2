```python
# Prints Hello, World!
def print_hello_world():
    print("Hello, World!")

print_hello_world()
```

```python
# Prints a greeting using the given name
def greet_name(name):
    print("Hello,", name)

greet_name("Mubarak")
```

```python
# Returns the square of a number
def square_number(n):
    return n * n

print(square_number(5))
```

```python
# Returns the sum of two numbers
def add_numbers(a, b):
    return a + b

print(add_numbers(3, 4))
```

```python
# Returns the maximum of two numbers
def max_number(a, b):
    if a > b:
        return a
    else:
        return b

print(max_number(7, 4))
```

```python
# Checks if a number is even or odd
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_or_odd(9))
```

```python
# Returns the length of a string
def string_length(s):
    return len(s)

print(string_length("Hello"))
```

```python
# Converts Celsius to Fahrenheit
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

print(celsius_to_fahrenheit(25))
```

```python
#Returns the average of a list of numbers
def average_list(numbers):
    return sum(numbers) / len(numbers)

print(average_list([2, 4, 6, 8]))
```

```python
#Counts vowels in a string
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count

print(count_vowels("Hello World"))
```

```python
#Returns a list with unique values
def unique_list(lst):
    return list(set(lst))

print(unique_list([1, 2, 2, 3, 4, 4]))
```

```python
#Checks if a word is a palindrome
def is_palindrome(word):
    return word == word[::-1]

print(is_palindrome("level"))
```

```python
#Calculates simple interest
def simple_interest(principal, rate, time):
    return (principal * rate * time) / 100

print(simple_interest(1000, 5, 2))
```

```python
#Function with default tax rate
def calculate_tax(amount, tax_rate=5):
    return amount * tax_rate / 100

print(calculate_tax(1000))
print(calculate_tax(1000, 10))
```

```python
#Returns the factorial of a number
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

print(factorial(5))
```

```python
#Returns the sum of any number of arguments
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4))
```

```python
#Returns the minimum value in a list
def min_list(lst):
    return min(lst)

print(min_list([5, 2, 9, 1]))
```

```python
#Converts minutes to hours and minutes
def minutes_to_hours(minutes):
    hours = minutes // 60
    mins = minutes % 60
    return hours, mins

print(minutes_to_hours(130))
```

```python
#Returns the area of a rectangle
def rectangle_area(length, width):
    return length * width

print(rectangle_area(5, 3))
```

```python
#Returns the largest number in a list
def max_list(lst):
    return max(lst)

print(max_list([5, 9, 2, 7]))
```

```python
#Checks if a number is positive, negative, or zero
def check_number(n):
    if n > 0:
        return "Positive"
    elif n < 0:
        return "Negative"
    else:
        return "Zero"

print(check_number(-4))
```

```python
#Returns the reverse of a string
def reverse_string(s):
    return s[::-1]

print(reverse_string("Python"))
```

```python
#Calculates the power of a number
def power(base, exponent):
    return base ** exponent

print(power(2, 3))
```

```python
#Returns grade based on score using default grading scale
def get_grade(score):
    if score >= 70:
        return "A"
    elif score >= 60:
        return "B"
    elif score >= 50:
        return "C"
    elif score >= 45:
        return "D"
    else:
        return "F"

print(get_grade(21))
```

```python
#Counts words in a sentence
def count_words(sentence):
    return len(sentence.split())

print(count_words("Muby is the most seniorman"))
```

```python
#Checks if a number is prime
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print(is_prime(17))
```

```python
#Returns the sum of digits of a number
def sum_digits(n):
    return sum(int(digit) for digit in str(abs(n)))

print(sum_digits(1234))
```

```python
#Applies a default discount to a list of prices
def apply_discount(prices, discount=10):
    return [price * (1 - discount/100) for price in prices]

print(apply_discount([100, 200, 300]))
```

```python
#Formats and returns full name from first and last name
def full_name(first, last):
    return f"{first} {last}"

print(full_name("John", "Yayi"))
```
