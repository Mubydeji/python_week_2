```python
# try: runs the risky code that might cause an error.
# execpt: Catches and handles the error if one occurs.
# else: Executes only if no exception occurs in try.  
# Finally: Runs regardless of what happens useful for cleanup tasks like closing files.
```

```python
#This is an error
print(name)
```

```python
# We fixing not defined errors

try:
    print(name)
except NameError:
    print ('name not defined')
```

```python
#The block of code will keep runing until it asked to stop, so you stop it when you enter 'q' in first or second
while True:
    first_num = input("\nFirst num: ")
    if first_num == "q":
        break
    second_num = input("Second_num: ")
    if second_num == "q":
        break
    try:
        answer = int(first_num) / int (second_num)
    except ZeroDivisionError:
        print("You can't divide by 0!")
    else:
        print(answer)
```

```python
#A function that take from the user a file name, 
#and caters for a siutation where the user might sent a wrong file name, if so it should not break, it should print the content of the file name

def read_file():
    filename = input("Enter the file name: ")

    try:
        with open(filename, "r") as f:
            print(f.read())
    except FileNotFoundError:
        print("file not found. check the file name and try again")
    except Exception as e:
        print("An unexpected error occurred:", e)

read_file()
```
